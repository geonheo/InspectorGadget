# Crowdsourcing Workflow with a Sample Image

<br>

```python
import InspectorG.crowdsourcing_workflow as cw
from InspectorG.utils import ImgViewer

org_path = './sample_images/original.png'
ImgViewer(org_path)
```

## Crowdsourcing with Inspector Gadget

Inspector Gadget can get marked image from crowd workers.
<br>
<br>

```python
marked_list = [f'./sample_images/marked_by_worker{i}.png' for i in range(5)]

for marked_image in marked_list:
    ImgViewer(marked_image)
```

## Extract bounding boxes from marked images

```python
group = cw.ExtractCandidate(org_path, marked_list, margin = 10, show = True) 
```

## Combine overlapping patterns and get outliers

```python
_, outliers = cw.CombinePatterns(org_path, group, save = True, show = True)
```

## Peer Review

```python
DTYPE = 'KSDD'

cw.PeerReview(org_path, outliers, save_outlier = False, defect_labels = [DTYPE])
```
